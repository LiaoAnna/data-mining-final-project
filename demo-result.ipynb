```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("./result.csv")

fig, ax = plt.subplots(figsize=(10, 6))
for method in df["Aug Method"].unique():
    subset = df[df["Aug Method"] == method]

    if method == "mask":
        # Highlight the specific point
        special_point = subset[subset["Aug Prob"] == 0.2]
        ax.scatter(special_point["Aug Prob"], special_point["Val Result"], label=method, marker="x", color="red", s=80)
    else:
        ax.scatter(subset["Aug Prob"], subset["Val Result"], label=method, marker="o", s=30)


# Labels and legend
ax.set_title("Comparison of Different Data Augmentation Methods", fontsize=14)
ax.set_xlabel("Augmentation Probability", fontsize=12)
ax.set_ylabel("Validation Result", fontsize=12)
ax.legend(title="Aug Method", bbox_to_anchor=(1.05, 1), loc="upper left")
ax.grid(True)

plt.tight_layout()
plt.show()
```
